# <h2 style="color:#2b6cb0;">2. Understanding Threat Intelligence</h2>

## <h3 style="color:#2f855a;">Definition and Key Concepts</h3>

Threat intelligence is the **structured process of collecting, cleaning/enriching, analyzing, and communicating data about cyber threats**. Its purpose is to help organizations make safer decisions by understanding attacker behaviors, identifying vulnerabilities, and predicting risks.

It applies to **any organization**â€”hospitals, financial institutions, public bodies, or private companies.

---

# <h3 style="color:#2f855a;">Simply put</h3>

> <strong>Threat intelligence = Collecting â†’ Enriching â†’ Analyzing â†’ Communicating</strong> threat information so an organization can anticipate attacks, strengthen defenses, and prevent operational disruptions and data breaches.

---

# <h3 style="color:#2f855a;">The Threat Intelligence Lifecycle</h3>

The threat intelligence lifecycle transforms **raw, noisy data** into **actionable, decision-ready intelligence**.
It consists of **four essential phases**:

---

## <h4 style="color:#805ad5;">1. Collection</h4>

**Goal:** Gather raw, unfiltered threat data.
**How:** Passive reconnaissance and open-source intelligence (OSINT).
**Methods:**

* Public data, web searches, threat feeds
* No interaction with the targetâ€™s internal systems
* Passive scanning

**Examples:**

* **Google Dorking** â†’ uncover exposed documents, login pages, error messages, or configuration files
* **Shodan** â†’ find exposed devices/ports


---

## <h4 style="color:#805ad5;">2. Enrichment & Processing</h4>

**Goal:** Clean and contextualize the collected dataâ€”similar to preparing ingredients before cooking.
**Methods:**

* Remove duplicates
* Validate inconsistent data
* Add technical or contextual details
* Link findings to known vulnerabilities, staff roles, systems, or external sources

**Example:**

* Cross-referencing emails with known breach databases
* Linking exposed subdomains to certificate transparency logs
* Mapping IP addresses to hosting providers

---

## <h4 style="color:#805ad5;">3. Analysis</h4>

**Goal:** Connect the dots and transform enriched data into meaningful insights.
**Methods:**

* Threat modeling
* Pattern recognition
* MITRE ATT&CK mapping
* Profiling threat actors and ransomware TTPs (tactics, techniques, procedures)

---

## <h4 style="color:#805ad5;">4. Dissemination</h4>

**Goal:** Communicate insights clearly to stakeholders who need them.
**Methods:**

* Executive summaries
* Alerts
* Detailed technical reports
* Recommendations for SOC teams, security leaders, or IT teams

**Example:**

* Summaries showing exposed credentials, vulnerable endpoints, and required mitigation steps
* Alerts about active exploitation of a critical CVE affecting the organization

---
