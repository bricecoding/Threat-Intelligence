# <h2 style="color:#2b6cb0;">2. Understanding Threat Intelligence</h2>

## <h3 style="color:#2f855a;">Definition and Key Concepts</h3>

Threat intelligence is the **structured process of collecting, cleaning/enriching, analyzing, and communicating data about cyber threats**. Its purpose is to help organizations make safer decisions by understanding attacker behaviors, identifying vulnerabilities, and predicting risks.

It applies to **any organization**‚Äîhospitals, financial institutions, public bodies, or private companies.

---

# <h3 style="color:#2f855a;">Simply put</h3>

> <strong>Threat intelligence = Collecting ‚Üí Enriching ‚Üí Analyzing ‚Üí Communicating</strong> threat information so an organization can anticipate attacks, strengthen defenses, and prevent operational disruptions and data breaches.

---

# <h3 style="color:#2f855a;">The Threat Intelligence Lifecycle</h3>

The threat intelligence lifecycle transforms **raw, noisy data** into **actionable, decision-ready intelligence**.
It consists of **four essential phases**:

---

## <h4 style="color:#805ad5;">1. Collection</h4>

**Goal:** Gather raw, unfiltered threat data.
**How:** Passive reconnaissance and open-source intelligence (OSINT).
**Methods:**

* Public data, web searches, threat feeds
* No interaction with the target‚Äôs internal systems
* Passive scanning

**Examples:**

* **Google Dorking** ‚Üí uncover exposed documents, login pages, error messages, or configuration files
* **Shodan** ‚Üí find exposed devices/ports


---

## <h4 style="color:#805ad5;">2. Enrichment & Processing</h4>

**Goal:** Clean and contextualize the collected data‚Äîsimilar to preparing ingredients before cooking.
**Methods:**

* Remove duplicates
* Validate inconsistent data
* Add technical or contextual details
* Link findings to known vulnerabilities, staff roles, systems, or external sources

**Example:**

* Cross-referencing emails with known breach databases
* Linking exposed subdomains to certificate transparency logs
* Mapping IP addresses to hosting providers

---

## <h4 style="color:#805ad5;">3. Analysis</h4>

**Goal:** Connect the dots and transform enriched data into meaningful insights.
**Methods:**

* Threat modeling
* Pattern recognition
* MITRE ATT&CK mapping
* Profiling threat actors and ransomware TTPs (tactics, techniques, procedures)

---

## <h4 style="color:#805ad5;">4. Dissemination</h4>

**Goal:** Communicate insights clearly to stakeholders who need them.
**Methods:**

* Executive summaries
* Alerts
* Detailed technical reports
* Recommendations for SOC teams, security leaders, or IT teams

**Example:**

* Summaries showing exposed credentials, vulnerable endpoints, and required mitigation steps
* Alerts about active exploitation of a critical CVE affecting the organization

---

# <h2 style="color:#2b6cb0;">3. Threat Intelligence Sources and Tools</h2>

## <h3 style="color:#2f855a;">Major Cyber Threat Intelligence Sources</h3>

Below are the six primary categories of threat intelligence sources used for real-world cybersecurity assessments:

---

### <h4 style="color:#805ad5;">‚Ä¢ Shodan</h4>

* Identifies exposed devices, misconfigured services, and outdated software.
* Helps assess the attack surface visible to the internet.

### <h4 style="color:#805ad5;">‚Ä¢ CISA KEV Catalog</h4>

* List of vulnerabilities actively exploited in the wild.
* Critical for prioritizing patching.

### <h4 style="color:#805ad5;">‚Ä¢ NVD (NIST Vulnerability Database)</h4>

* Authoritative CVE database.
* Provides severity, impact, and technical details.

### <h4 style="color:#805ad5;">‚Ä¢ VirusTotal</h4>

* Aggregates antivirus results for files, URLs, and IP addresses.
* Useful for malware reputation and behavioral analysis.

### <h4 style="color:#805ad5;">‚Ä¢ HaveIBeenPwned (HIBP)</h4>

* Checks emails against known data breaches.
* Helps assess phishing and credential reuse risks.

### <h4 style="color:#805ad5;">‚Ä¢ NHS Digital Cyber Alerts</h4>

* Sector-specific advisories for the NHS.
* Equivalent sector feeds exist for finance, energy, aviation, etc.

---

# <h3 style="color:#2f855a;">Practical Tools for Threat Intelligence</h3>

### <h4 style="color:#805ad5;">‚Ä¢ theHarvester</h4>

Collects publicly available emails, subdomains, and IPs. Simulates passive attacker reconnaissance.

### <h4 style="color:#805ad5;">‚Ä¢ Hunter.io</h4>

Identifies corporate email patterns. Helps map organizational structure and phishing risk.

### <h4 style="color:#805ad5;">‚Ä¢ crt.sh</h4>

Searches certificate transparency logs.
Useful for discovering forgotten subdomains and third-party dependencies.

### <h4 style="color:#805ad5;">‚Ä¢ Whois</h4>

Provides domain ownership details, registrar, and infrastructure information.

### <h4 style="color:#805ad5;">‚Ä¢ Google Dorking</h4>

Uses advanced Google operators to reveal sensitive files, login pages, error logs, and misconfigurations.

Useful for:

* Discovering exposed **PDFs**, **docs**, **config files**
* Finding **error messages**, **internal URLs**, and **stack traces**
* Uncovering **supply chain exposure** through third-party domain references

### <h4 style="color:#805ad5;">‚Ä¢ Shodan CLI/Web</h4>

Enables querying of public-facing internet assets.
Useful for mapping services and detecting high-risk exposures.
